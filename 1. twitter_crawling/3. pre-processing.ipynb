```python
DATA_DIR = "data/"
DATA_DOC = os.path.join(DATA_DIR, "data.csv")
CLEANED_DATA = os.path.join(DATA_DIR, "cleaned_7th_Apr.csv")
```

```python
def csv_to_df(filepath):
    """
    Convert CSV file to dataframe
    Parameter: 
        filepath: path of csv file to be read
    Return: pandas dataframe
    """
    df1 = pd.read_csv(filepath)
    return df1

def df_to_csv(df, filepath):
    """
    Convert dataframe to CSV file
    Parameters:
        df: dataframe
        filepath: path of csv file to write to
    """
    df.to_csv(filepath, encoding='utf-8', index=False)
```

```python
df = csv_to_df(DATA_DOC)
data = df[['text', 'entities', 'user', 'geo']]
data
```

```python
def clean_tweets(text):    
    # print("before: " + text)
    # text = re.sub("'", "", text) # to avoid removing contractions in english
    text = re.sub("@[A-Za-z0-9_]+","", text) # remove @mentions
    text = re.sub(r"#", "", text)  # Normalise hashtags by dropping '#'
    text = re.sub(r'http\S+', '', text) # remove links
    text = re.sub(r"www.\S+", "", text) # remove links
    text = re.sub("[^A-Za-z0-9./\\\()!?\'\"]"," ", text) # filter non-alphanumeric characters
    text = re.sub(" +", " ", text) # remove trailling spaces
    text = re.sub(r"\r\n", " ", text) # remove newlines

    return text
    # ()!?./\
    # print("AFTER: " + text)
```

```python
def recursive_find(dict_, key):
    if not isinstance(dict_, dict):
        return dict_
    if dict_.get(key) is not None:
        return dict_[key]

    for value in dict_.values():
        for new_dict in value:
            output = recursive_find(new_dict, key)
            if output is not None:
                return output

    return None


def func(x, key):
    if x is np.nan or isinstance(ast.literal_eval(x), str):
        return x
    return recursive_find(ast.literal_eval(x), key)
    
```

```python
def extract_tags(text):
    try: 
        dict = ast.literal_eval(text)

        taglist = []

        for i in dict['hashtags']:
            taglist.append(i['text'])
    except SyntaxError:
        taglist = []
    
    return taglist
```

```python
def extract_userID(text):
    try:
        dict = ast.literal_eval(text)

        userID = dict['id']
    except SyntaxError:
        userID = ""

    return userID
```
